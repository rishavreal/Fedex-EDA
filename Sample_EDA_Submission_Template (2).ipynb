## 3. ***Data Wrangling***

### Data Wrangling Code

````python
#removing duplicate value
df.drop_duplicates(inplace=True)
````

````python
df.fillna(df.mean(numeric_only=True), inplace=True) #filling missing numerical values with mean
df.fillna(df.mode().iloc[0],inplace=True) # filling categorical missing values with mode
````

````python
df.isnull().sum()
````

### What all manipulations have you done and insights you found?

- Removed duplicates if any
- Handled missing values by putting mean for numerical values and mode for categorical values.

````python
plt.figure(figsize=(15,8))
for x in df['Fulfill Via'].unique():
    df1= df[df['Fulfill Via']==x]
    plt.plot(df1.groupby('Country')['Fulfill Via'].count(), label= x)
plt.xlabel('Country')
plt.xticks(rotation=90)
plt.ylabel('Methods')
plt.title('Different methods of shipment vs country')
plt.legend()
plt.grid(True)
plt.show()
````

##### 1. Why did you pick the specific chart?

This chart clearly shows the different shipment methods used by different countries and indicates whether the product is from direct drop or from RDC

##### 2. What is/are the insight(s) found from the chart?

Most of the less developed countries like congo , Nigeria , uganda etc.,  have shipments from the RDC while south africa , vietnam have most of the shipment from direct drop.
